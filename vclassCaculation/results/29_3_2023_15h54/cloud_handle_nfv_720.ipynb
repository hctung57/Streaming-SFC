```python
for i in range(0, number_of_reptition):
    file_name = "data_prom_sfc_1_720_repeat_time_{}.csv".format(
        i)

    ram_match_av = []
    cpu_match_av = []
    bandwidth_match_av = []
    bandwidth_transmit_match_av = []
    bandwidth_reiceive_match_av = []
    time_match_av = []

    with open(file_name, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for row in spamreader:
            if "match-av" in row[5]:
                cpu_match_av.append(float(row[1]))
                ram_match_av.append(float(row[2])/10**6)
                bandwidth_transmit_match_av.append(float(row[4]))
                bandwidth_reiceive_match_av.append(float(row[3]))
                time_match_av.append(float(row[0]))

        average_cpu_match_av.append(np.mean(cpu_match_av))
        average_ram_match_av.append(np.mean(ram_match_av))
        average_bandwidth_transmit_match_av.append(
            (bandwidth_transmit_match_av[-1]-bandwidth_transmit_match_av[0])/(time_match_av[-1]-time_match_av[0]))
        average_bandwidth_reiceive_match_av.append(
            (bandwidth_reiceive_match_av[-1]-bandwidth_reiceive_match_av[0])/(time_match_av[-1]-time_match_av[0]))


ram_match_av_mean=np.mean(average_ram_match_av)
cpu_match_av_mean=np.mean(average_cpu_match_av)
bandwidth_transmit_match_av_mean=np.mean(average_bandwidth_transmit_match_av)
bandwidth_reiceive_match_av_mean=np.mean(average_bandwidth_reiceive_match_av)

ram_match_av_std=np.std(average_ram_match_av)
cpu_match_av_std=np.std(average_cpu_match_av)
bandwidth_transmit_match_av_std=np.std(average_bandwidth_transmit_match_av)
bandwidth_reiceive_match_av_std=np.std(average_bandwidth_reiceive_match_av)
```
