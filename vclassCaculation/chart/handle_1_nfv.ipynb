```python
import matplotlib.pyplot as plot
import csv
import numpy as np
```

```python
average_ram_source_streaming = []
average_cpu_source_streaming = []
average_bandwidth_transmit_source_streaming = []
average_bandwidth_reiceive_source_streaming = []
average_time_source_streaming = []

average_ram_face_detection = []
average_cpu_face_detection = []
average_bandwidth_transmit_face_detection = []
average_bandwidth_reiceive_face_detection = []
average_time_face_detection = []

average_ram_background_blur = []
average_cpu_background_blur = []
average_bandwidth_transmit_background_blur = []
average_bandwidth_reiceive_background_blur = []
average_time_background_blur = []

average_ram_transcoder = []
average_cpu_transcoder = []
average_bandwidth_transmit_transcoder = []
average_bandwidth_reiceive_transcoder = []
average_time_transcoder = []

average_ram_match_av = []
average_cpu_match_av = []
average_bandwidth_transmit_match_av = []
average_bandwidth_reiceive_match_av = []
average_time_match_av = []

average_ram_noise_reduction = []
average_cpu_noise_reduction = []
average_bandwidth_transmit_noise_reduction = []
average_bandwidth_reiceive_noise_reduction = []
average_time_noise_reduction = []
```

```python
for i in range(0, 10):
    file_name = "../results/14_3_2023_0h22/data_prom_sfc_1_14_3_2023_0h22_repeat_time_{}.csv".format(
        i)

    ram_source_streaming = []
    cpu_source_streaming = []
    bandwidth_source_streaming = []
    bandwidth_transmit_source_streaming = []
    bandwidth_reiceive_source_streaming = []
    time_source_streaming = []

    ram_noise_reduction = []
    cpu_noise_reduction = []
    bandwidth_noise_reduction = []
    bandwidth_transmit_noise_reduction = []
    bandwidth_reiceive_noise_reduction = []
    time_noise_reduction = []

    ram_match_av = []
    cpu_match_av = []
    bandwidth_match_av = []
    bandwidth_transmit_match_av = []
    bandwidth_reiceive_match_av = []
    time_match_av = []

    with open(file_name, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for row in spamreader:
            if "source-streaming" in row[5]:
                cpu_source_streaming.append(float(row[1]))
                ram_source_streaming.append(float(row[2])/10**6)
                bandwidth_transmit_source_streaming.append(float(row[4]))
                bandwidth_reiceive_source_streaming.append(float(row[3]))
                time_source_streaming.append(float(row[0]))
            if "noise-suppress" in row[5]:
                cpu_noise_reduction.append(float(row[1]))
                ram_noise_reduction.append(float(row[2])/10**6)
                bandwidth_transmit_noise_reduction.append(float(row[4]))
                bandwidth_reiceive_noise_reduction.append(float(row[3]))
                time_noise_reduction.append(float(row[0]))
            if "match-av" in row[5]:
                cpu_match_av.append(float(row[1]))
                ram_match_av.append(float(row[2])/10**6)
                bandwidth_transmit_match_av.append(float(row[4]))
                bandwidth_reiceive_match_av.append(float(row[3]))
                time_match_av.append(float(row[0]))

        average_cpu_source_streaming.append(np.mean(cpu_source_streaming))
        average_ram_source_streaming.append(np.mean(ram_source_streaming))
        average_bandwidth_transmit_source_streaming.append(
            (bandwidth_transmit_source_streaming[-1]-bandwidth_transmit_source_streaming[0])/(time_source_streaming[-1]-time_source_streaming[0]))
        average_bandwidth_reiceive_source_streaming.append(
            (bandwidth_reiceive_source_streaming[-1]-bandwidth_reiceive_source_streaming[0])/(time_source_streaming[-1]-time_source_streaming[0]))

        average_cpu_noise_reduction.append(np.mean(cpu_noise_reduction))
        average_ram_noise_reduction.append(np.mean(ram_noise_reduction))
        average_bandwidth_transmit_noise_reduction.append(
            (bandwidth_transmit_noise_reduction[-1]-bandwidth_transmit_noise_reduction[0])/(time_noise_reduction[-1]-time_noise_reduction[0]))
        average_bandwidth_reiceive_noise_reduction.append(
           (bandwidth_reiceive_noise_reduction[-1]-bandwidth_reiceive_noise_reduction[0])/(time_noise_reduction[-1]-time_noise_reduction[0]))

        average_cpu_match_av.append(np.mean(cpu_match_av))
        average_ram_match_av.append(np.mean(ram_match_av))
        average_bandwidth_transmit_match_av.append(
            (bandwidth_transmit_match_av[-1]-bandwidth_transmit_match_av[0])/(time_match_av[-1]-time_match_av[0]))
        average_bandwidth_reiceive_match_av.append(
            (bandwidth_reiceive_match_av[-1]-bandwidth_reiceive_match_av[0])/(time_match_av[-1]-time_match_av[0]))

ram_source_streaming_mean=np.mean(average_ram_source_streaming)
cpu_source_streaming_mean=np.mean(average_cpu_source_streaming)
bandwidth_transmit_source_streaming_mean=np.mean(
    average_bandwidth_transmit_source_streaming)
bandwidth_reiceive_source_streaming_mean=np.mean(
    average_bandwidth_reiceive_source_streaming)

ram_noise_reduction_mean=np.mean(average_ram_noise_reduction)
cpu_noise_reduction_mean=np.mean(average_cpu_noise_reduction)
bandwidth_transmit_noise_reduction_mean=np.mean(
    average_bandwidth_transmit_noise_reduction)
bandwidth_reiceive_noise_reduction_mean=np.mean(
    average_bandwidth_reiceive_noise_reduction)

ram_match_av_mean=np.mean(average_ram_match_av)
cpu_match_av_mean=np.mean(average_cpu_match_av)
bandwidth_transmit_match_av_mean=np.mean(average_bandwidth_transmit_match_av)
bandwidth_reiceive_match_av_mean=np.mean(average_bandwidth_reiceive_match_av)

ram_source_streaming_std=np.std(average_ram_source_streaming)
cpu_source_streaming_std=np.std(average_cpu_source_streaming)
bandwidth_transmit_source_streaming_std=np.std(
    average_bandwidth_transmit_source_streaming)
bandwidth_reiceive_source_streaming_std=np.std(
    average_bandwidth_reiceive_source_streaming)

ram_noise_reduction_std=np.std(average_ram_noise_reduction)
cpu_noise_reduction_std=np.std(average_cpu_noise_reduction)
bandwidth_transmit_noise_reduction_std=np.std(
    average_bandwidth_transmit_noise_reduction)
bandwidth_reiceive_noise_reduction_std=np.std(
    average_bandwidth_reiceive_noise_reduction)

ram_match_av_std=np.std(average_ram_match_av)
cpu_match_av_std=np.std(average_cpu_match_av)
bandwidth_transmit_match_av_std=np.std(average_bandwidth_transmit_match_av)
bandwidth_reiceive_match_av_std=np.std(average_bandwidth_reiceive_match_av)
```

```python
# CPU in cloud
data_receive = [bandwidth_reiceive_source_streaming_mean,
                bandwidth_reiceive_match_av_mean, bandwidth_reiceive_noise_reduction_mean]
data_transmit = [bandwidth_transmit_source_streaming_mean,
                 bandwidth_transmit_match_av_mean, bandwidth_transmit_noise_reduction_mean]
# error = [bandwidth_reiceive_source_streaming_std,
#           bandwidth_transmit_source_streaming_std]
state = ["source streaming", "match av", "noise reduction"]
fig1, ax1 = plot.subplots()
x_pos = x_pos = np.arange(len(state))
width = 0.4
ax1.bar(x_pos, data_receive, width=width, align='center',
        alpha=0.9, ecolor='black', capsize=10)
ax1.bar(x_pos+width, data_transmit, width=width, align='center',
        alpha=0.9, ecolor='black', capsize=10)
ax1.set_title('CPU in cloud')
ax1.set_ylabel("%")
ax1.set_xticks(x_pos)
ax1.set_xticklabels(state)
plot.show()
```
